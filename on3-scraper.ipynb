```python
# ID HTML elements in on3 pg

# create a headless firefox driver
options = Options()
options.add_argument("--headless")
driver = webdriver.Firefox(options=options)
driver.get("https://www.on3.com/transfer-portal/wire/womens-basketball/2025/")
time.sleep(5)

# links = driver.find_elements(By.TAG_NAME, "a")
# print(links)

# # get href names only if "rivals" in href
# hrefNames = [links[i].get_attribute("href") for i in range(len(links))]

# # hrefNames = [links[i].get_attribute("href") for i in range(len(links))]

# for name in hrefNames:
#     if name and "rivals" in name:
#         print(name)

# Get all ols
ols = driver.find_elements(By.TAG_NAME, "ol")
# filter by class
for ol in ols:
    if "TransferPortal" in ol.get_attribute("class"):
        list_items = ol.find_elements(By.TAG_NAME, "li")
        for item in list_items:
            print(item.text)

# close driver
driver.quit()
```
