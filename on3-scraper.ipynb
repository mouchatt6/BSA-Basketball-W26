# Find all API Calls #

```python
import requests
import re

# Get the main JavaScript bundle
response = requests.get('https://www.on3.com/transfer-portal/wire/womens-basketball/2025')
html = response.text

# Find script tags
scripts = re.findall(r'<script src="([^"]+)"', html)

for script_url in scripts:
    if not script_url.startswith('http'):
        script_url = 'https://www.on3.com' + script_url
    
    try:
        js = requests.get(script_url).text
        
        # Search for API endpoints
        endpoints = re.findall(r'api\.on3\.com[^"\'}\s]+.*$', js)
        for endpoint in set(endpoints):
            print(f"Found: {endpoint}")
            
    except:
        pass
```
