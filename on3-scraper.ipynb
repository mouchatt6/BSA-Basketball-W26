```python
import numpy as np
import pandas as pd
import polars as pl
import scrapy
import selenium as sl
from selenium import webdriver  
from selenium.webdriver.common.by import By  
import time  
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from tqdm.notebook import tqdm
import os
import requests 
```

```python
# ID HTML elements in on3 pg

# create a headless firefox driver
options = Options()
options.add_argument("--headless")
driver = webdriver.Firefox(options=options)
driver.get("https://www.on3.com/transfer-portal/wire/womens-basketball/2025/")
time.sleep(5)

# Get all ols
ols = driver.find_elements(By.TAG_NAME, "ol")
# filter by class
for ol in ols:
    if "TransferPortal" in ol.get_attribute("class"):
        list_items = ol.find_elements(By.TAG_NAME, "li")
        # find a tags inside li and get text
        for item in list_items:
            a_tags = item.find_elements(By.TAG_NAME, "a")
            hrefs = [a.get_attribute("href") for a in a_tags]
            for i, a_tag in enumerate(a_tags):
                href = hrefs[i]
                if href and "rivals" in href:
                    name = a_tag.text
                    if name:
                      print(name, href)
                if href and "college" in href:
                    # get school from href link
                    school = href.split("college")[1].split("/")[1].replace("-", " ")
                    if "committedLogoContainer" in a_tag.get_attribute("class"):
                        print(school, "- Committed")
                    else:
                        print(school, "- left")
                    # print(href)

            # get Position
            # # get text inside span
            # span_tags = item.find_elements(By.TAG_NAME, "span")
            # for span_tag in span_tags:
            #     pos = span_tag.text
            #     if pos:
            #         print(pos)

            # School from and to
            # get links with college in hre
                

# close driver
driver.quit()
```
